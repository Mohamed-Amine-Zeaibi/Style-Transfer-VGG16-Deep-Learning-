```python
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras.applications import VGG19
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, Add, ReLU
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg19 import preprocess_input
```

```python
content_path = "./content/content.jpeg"
vangogh_path = "./style/vangogh.jpg"
content_img = Image.open(content_path).resize((256,256))
vangogh_img = Image.open(vangogh_path).resize((256,256))

content_array = np.array(content_img, dtype=np.float32)
vangogh_array = np.array(vangogh_img, dtype=np.float32)

content_array = preprocess_input(content_array * 255.0)
vangogh_array = preprocess_input(vangogh_array * 255.0)

content = np.expand_dims(content_array, axis=0)
vangogh = np.expand_dims(vangogh_array, axis=0)
```

```python
input_img = Input(shape=(256,256,3))

x = Conv2D(32, (9,9), padding="same", activation="relu", kernel_initializer='he_normal')(input_img)
x = Conv2D(64, (9,9), padding="same", activation="relu", kernel_initializer='he_normal')(x)
x = Conv2D(128, (3,3), strides=2, padding="same", activation="relu", kernel_initializer='he_normal')(x)
x = Conv2D(128, (3,3), strides=2, padding="same", activation="relu", kernel_initializer='he_normal')(x)

shortcut = x
for _ in range(5):
    res = Conv2D(128, (3,3), padding="same", activation="relu", kernel_initializer='he_normal')(x)
    res = Conv2D(128, (3,3), padding="same", kernel_initializer='he_normal')(res)
    x = Add()([res, shortcut])
    x = ReLU()(x)
    shortcut = x
from tensorflow.keras.layers import UpSampling2D
x = UpSampling2D(size=(2,2))(x)
x = Conv2D(64, (3,3), padding="same", activation="relu")(x)
x = UpSampling2D(size=(2,2))(x)
x = Conv2D(32, (3,3), padding="same", activation="relu")(x)
output_img = Conv2D(3, (9,9), padding="same", activation="sigmoid")(x)
generator = Model(inputs=input_img, outputs=output_img)
generator.summary()
```

```python
import matplotlib.pyplot as plt

# Charger et prparer les images
content_img = Image.open("./content/content.jpeg").resize((512, 512))
style_img = Image.open("./style/vangogh.jpg").resize((512, 512))

content = np.array(content_img, dtype=np.float32) / 255.0
style = np.array(style_img, dtype=np.float32) / 255.0

# Afficher les images originales
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(content)
plt.title("Image Content")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(style)
plt.title("Image Style")
plt.axis('off')
plt.show()

# Image  optimiser (commence par une copie du content)
generated = tf.Variable(content[np.newaxis, ...])

# Extracteur VGG19
vgg = tf.keras.applications.VGG19(include_top=False, weights='imagenet')
vgg.trainable = False

# Couches pour style et content
style_layers = ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1', 'block5_conv1']
content_layer = 'block4_conv2'

outputs = [vgg.get_layer(name).output for name in style_layers + [content_layer]]
feature_extractor = tf.keras.Model(inputs=vgg.input, outputs=outputs)

# Gram matrix pour style
def gram_matrix(input_tensor):
    result = tf.linalg.einsum('bijc,bijd->bcd', input_tensor, input_tensor)
    input_shape = tf.shape(input_tensor)
    num_locations = tf.cast(input_shape[1] * input_shape[2], tf.float32)
    return result / num_locations

# Get features
def get_features(image):
    preprocessed = tf.keras.applications.vgg19.preprocess_input(image * 255.0)
    return feature_extractor(preprocessed)

# Targets
style_targets = get_features(style[np.newaxis, ...])
content_target = get_features(content[np.newaxis, ...])[-1]

style_grams = [gram_matrix(style_feat) for style_feat in style_targets[:len(style_layers)]]

# Poids
style_weight = 1e4
content_weight = 1e0

# Optimisation
optimizer = tf.optimizers.Adam(learning_rate=0.01)

for epoch in range(100):
    with tf.GradientTape() as tape:
        gen_features = get_features(generated)
        
        # Content loss
        gen_content = gen_features[-1]
        content_loss = tf.reduce_mean(tf.square(gen_content - content_target))
        
        # Style loss
        style_loss = 0
        for i in range(len(style_layers)):
            gen_gram = gram_matrix(gen_features[i])
            style_loss += tf.reduce_mean(tf.square(gen_gram - style_grams[i]))
        
        total_loss = content_weight * content_loss + style_weight * style_loss
    
    gradients = tape.gradient(total_loss, [generated])
    optimizer.apply_gradients(zip(gradients, [generated]))
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss.numpy()}")
        
        # Afficher rsultat
        img = generated[0].numpy()
        img = np.clip(img, 0, 1)
        plt.figure(figsize=(5, 5))
        plt.imshow(img)
        plt.title(f"Epoch {epoch}")
        plt.axis('off')
        plt.show()

print("Termin!")
```
